```
df = pd.read_csv('allbp-data.csv')
test = pd.read_csv('allbp-test.csv')
```

```
df['TBG'].unique()
```

```
df.drop(columns='TBG', inplace = True)
```

```
for col in df.columns:
    df.drop(df[df[col] == '?'].index, inplace = True)
```

```
df.describe()
```

```
df.head()
```

```
for col in df.columns:
    if not col in ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']:
      median = df[col].describe().top
      df[col] = df[col].apply(lambda x: median if x == '?' else x)
```

```
for col in df.columns:
    if col in ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']:
      df[col] = df[col].apply(lambda x: 0 if x == '?' else x)

for col in df.columns:
    if col in ['age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI']:
      df[col] = df[col].apply(lambda x: 0 if x == '?' else x)
```

```
X = df.drop(columns='class')
```

```
X.head()
```

```
for col in X.columns:
  if not col in ['sex','age', 'TSH', 'T3', 'TT4', 'T4U', 'FTI', 'TBG', 'referral source']:
    X[col] = X[col].apply(lambda x: 1 if x == 'f' else '0')

X['sex'] = X['sex'].apply(lambda x: 1 if x == 'F' else '0')
```

```
X['referral source'].unique()
```

```
X['referral source'] = X['referral source'].apply(lambda x: 0 if x == 'SVHC' else (1 if x == 'other' else (2 if x == 'SVI' else (3 if x == 'STMW' else 4))))
```

```
X.head()
```

```
y = df['class']
```

```
y = y.apply(lambda x: 0 if x == 'negative' else (1 if x == 'increased binding protein' else 2))
```

```
import numpy as np
```

```
y.head()
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = 0.3, shuffle=False)
```

# Modelling

```
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import re
import spacy
from gensim.parsing.preprocessing import remove_stopwords
import gensim

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import RepeatedStratifiedKFold
from scipy.stats import loguniform

 
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```
nb = MultinomialNB()
nb.fit(X_train, y_train)
nb_pred = nb.predict(X_test)
accuracy_score(y_test, nb_pred)
```

```
model = MultinomialNB()
space = dict()
space['alpha'] = [i for i in range(1, 1000, 10)]
# define search
search = GridSearchCV(model, space, scoring='accuracy', n_jobs=-1)
# execute search
result = search.fit(X_train, y_train)
# summarize result
print('Best Score: %s' % result.best_score_)
print('Best Hyperparameters: %s' % result.best_params_)
```

```
print("Classification Report:")
logistic_class_report = classification_report(y_test, result.predict(X_test))
print(logistic_class_report)
```
